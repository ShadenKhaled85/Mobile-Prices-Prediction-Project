# Data Cleaning

### Checking missing or wrong values

```python
# Checking unrealistic values values
print(f"px_height zero or less count: {prices[prices['px_height'] <= 0].shape[0]}")
print(f"px_width zero or less count: {prices[prices['px_width'] <= 0].shape[0]}")
print(f"Ram zero or less count: {prices[prices['ram'] <= 0].shape[0]}")
print(f"Battery Power zero or less count: {prices[prices['battery_power'] <= 0].shape[0]}")
print(f"Mobile weight zero or less count: {prices[prices['mobile_wt'] <= 0].shape[0]}")
print(f"Screen width zero or less count: {prices[prices['sc_w'] <= 0].shape[0]}")
print(f"Screen hight zero or less count: {prices[prices['sc_h'] <= 0].shape[0]}")
```

```python
#handling the unrealsitic data with imputing
# List of columns with unrealistic values
unrealistic_columns = ['px_height', 'px_width', 'ram', 'battery_power', 'mobile_wt', 'sc_w', 'sc_h']

# Imputing zero or negative values with the median of each column
for col in unrealistic_columns:
    median_value = prices[col].median()
    prices[col] = prices[col].apply(lambda x: median_value if x <= 0 else x)
    
# Checking data shape after imputation
print(f"Data shape after imputation: {prices.shape}")
```

```python
# Identifying duplicate rows
duplicate_rows = prices[prices.duplicated()]
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")

# Display the duplicate rows
duplicate_rows.head()
```

```python
#final checkkk
print(f"px_height zero or less count: {prices[prices['px_height'] <= 0].shape[0]}")
print(f"px_width zero or less count: {prices[prices['px_width'] <= 0].shape[0]}")
print(f"Ram zero or less count: {prices[prices['ram'] <= 0].shape[0]}")
print(f"Battery Power zero or less count: {prices[prices['battery_power'] <= 0].shape[0]}")
print(f"Mobile weight zero or less count: {prices[prices['mobile_wt'] <= 0].shape[0]}")
print(f"Screen width zero or less count: {prices[prices['sc_w'] <= 0].shape[0]}")
print(f"Screen height zero or less count: {prices[prices['sc_h'] <= 0].shape[0]}")
```

### Feature Engnieering

```python
# Feature engineering: Pixel area
prices['px_area'] = prices['px_height'] * prices['px_width']
```

```python
# Binning RAM into 4 categories using quartiles
prices['ram_bin'] = pd.qcut(prices['ram'], q=4, labels=[0, 1, 2, 3])
```

```python
# One-hot encoding for 'n_cores'
prices = pd.get_dummies(prices, columns=['n_cores'], drop_first=True)
```

```python
#Scaling the continuous features is essential to ensure that models like SVM, logistic regression, or k-NN perform optimally
from sklearn.preprocessing import StandardScaler

# Defining continuous numerical columns for scaling
numerical_cols = ['battery_power', 'ram', 'mobile_wt', 'px_height', 'px_width', 'px_area']

# Scaling the continuous features
scaler = StandardScaler()
prices[numerical_cols] = scaler.fit_transform(prices[numerical_cols])
```

```python
# Function to handle outliers using IQR
def remove_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Apply to features with high variability
prices = remove_outliers(prices, 'ram')
prices = remove_outliers(prices, 'px_height')
```

```python
prices.to_csv('final_data.csv',index=False)
```

# Preparing the Data

```python
from sklearn.model_selection import train_test_split

# Defining X and y
X = prices.drop(columns=['price_range'])
y = prices['price_range']

# Splitting into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression Classifier
lr_classifier = LogisticRegression(max_iter=6000, random_state=42)

# Train the model
lr_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred_lr = lr_classifier.predict(X_test)
```

# Evaluation Metrics

### Accuracy 

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print("Logistic Regression Accuracy:", accuracy_lr)
```

### Precision, Recall, and F1-Score

```python
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_lr))
```

### Confusion Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred_lr), annot=True, fmt="d", cmap="Blues")
plt.title("Logistic Regression Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# SVM Model

```python
# Import SVM Classifier
from sklearn.svm import SVC
# Initialize the Support Vector Classifier
svm_model = SVC(kernel='linear', random_state=42)
# Train the model
svm_model.fit(X_train, y_train)
# Make predictions on the test set
y_pred_svm = svm_model.predict(X_test)
```

```python
# SVM Accuracy
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print("SVM Accuracy:", accuracy_svm)
```

```python
# SVM Classification Report
print("SVM Classification Report:\n", classification_report(y_test, y_pred_svm))

# SVM Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred_svm), annot=True, fmt="d", cmap="Greens")
plt.title("SVM Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# Comparison of Models

```python
# Comparison of Models
print("\nComparison of Models:")
print(f"Logistic Regression Accuracy: {accuracy_lr}")
print(f"SVM Accuracy: {accuracy_svm}")

# Confusion matrices will already be displayed by the heatmaps
```

```python
from sklearn.metrics import precision_recall_fscore_support

# Extract precision, recall, f1-score for both models
metrics_lr = precision_recall_fscore_support(y_test, y_pred_lr, average='weighted')
metrics_svm = precision_recall_fscore_support(y_test, y_pred_svm, average='weighted')

# Creating a DataFrame for easier plotting
comparison_df = pd.DataFrame({
    'Metric': ['Precision', 'Recall', 'F1-Score'],
    'Logistic Regression': [metrics_lr[0], metrics_lr[1], metrics_lr[2]],
    'SVM': [metrics_svm[0], metrics_svm[1], metrics_svm[2]]
})

# Plotting precision, recall, and F1-score
comparison_df.plot(x='Metric', kind='bar', figsize=(8, 6), color=['blue', 'green'])
plt.title('Precision, Recall, and F1-Score Comparison')
plt.ylabel('Score')
plt.ylim([0.5, 1.0])
plt.show()
```
