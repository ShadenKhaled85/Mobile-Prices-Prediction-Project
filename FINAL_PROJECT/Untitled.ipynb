# Data Cleaning

### Checking missing or wrong values

```python
# Checking unrealistic values values
print(f"px_height zero or less count: {prices[prices['px_height'] <= 0].shape[0]}")
print(f"px_width zero or less count: {prices[prices['px_width'] <= 0].shape[0]}")
print(f"Ram zero or less count: {prices[prices['ram'] <= 0].shape[0]}")
print(f"Battery Power zero or less count: {prices[prices['battery_power'] <= 0].shape[0]}")
print(f"Mobile weight zero or less count: {prices[prices['mobile_wt'] <= 0].shape[0]}")
print(f"Screen width zero or less count: {prices[prices['sc_w'] <= 0].shape[0]}")
print(f"Screen hight zero or less count: {prices[prices['sc_h'] <= 0].shape[0]}")
```

```python
#handling the unrealsitic data with imputing
# List of columns with unrealistic values
unrealistic_columns = ['px_height', 'px_width', 'ram', 'battery_power', 'mobile_wt', 'sc_w', 'sc_h']

# Imputing zero or negative values with the median of each column
for col in unrealistic_columns:
    median_value = prices[col].median()
    prices[col] = prices[col].apply(lambda x: median_value if x <= 0 else x)
    
# Checking data shape after imputation
print(f"Data shape after imputation: {prices.shape}")
```

```python
# Identifying duplicate rows
duplicate_rows = prices[prices.duplicated()]
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")

# Display the duplicate rows
duplicate_rows.head()
```

```python
#final checkkk
print(f"px_height zero or less count: {prices[prices['px_height'] <= 0].shape[0]}")
print(f"px_width zero or less count: {prices[prices['px_width'] <= 0].shape[0]}")
print(f"Ram zero or less count: {prices[prices['ram'] <= 0].shape[0]}")
print(f"Battery Power zero or less count: {prices[prices['battery_power'] <= 0].shape[0]}")
print(f"Mobile weight zero or less count: {prices[prices['mobile_wt'] <= 0].shape[0]}")
print(f"Screen width zero or less count: {prices[prices['sc_w'] <= 0].shape[0]}")
print(f"Screen height zero or less count: {prices[prices['sc_h'] <= 0].shape[0]}")
```

### Feature Engnieering

```python
# Feature engineering: Pixel area
prices['px_area'] = prices['px_height'] * prices['px_width']
```

```python
# Binning RAM into 4 categories using quartiles
prices['ram_bin'] = pd.qcut(prices['ram'], q=4, labels=[0, 1, 2, 3])
```

```python
# One-hot encoding for 'n_cores'
prices = pd.get_dummies(prices, columns=['n_cores'], drop_first=True)
```

```python
#Scaling the continuous features is essential to ensure that models like SVM, logistic regression, or k-NN perform optimally
from sklearn.preprocessing import StandardScaler

# Defining continuous numerical columns for scaling
numerical_cols = ['battery_power', 'ram', 'mobile_wt', 'px_height', 'px_width', 'px_area']

# Scaling the continuous features
scaler = StandardScaler()
prices[numerical_cols] = scaler.fit_transform(prices[numerical_cols])
```

```python
# Function to handle outliers using IQR
def remove_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Apply to features with high variability
prices = remove_outliers(prices, 'ram')
prices = remove_outliers(prices, 'px_height')
```
